```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np

# 1. Resmi yükle
img = image.load_img("cat.jpg", target_size=(224, 224))  # resmi 224x224'e yeniden boyutlandırıyoruz

# 2. NumPy array'e çevir
img_array = image.img_to_array(img)

# 3. Normalizasyon (0-1 aralığına getiriyoruz)
img_array = img_array / 255.0

# 4. Model girişine uygun şekle getir (batch ekle)
img_array = np.expand_dims(img_array, axis=0)

print("Resim array boyutu:", img_array.shape)

# 5. Görselleştir
plt.imshow(img)
plt.title("Yüklenen Kedi Resmi")
plt.axis("off")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

# 1. Tek resmimizi yükle
img = load_img("cat.jpg")       # resmin yolu
x = img_to_array(img)           # numpy array'e çevir
x = x.reshape((1,) + x.shape)   # batch boyutu ekle

# 2. Data augmentation tanımla
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 3. Klasör oluştur (çıkış klasörü)
save_dir = "augmented_images"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 4. Augmented görüntüleri üret ve kaydet
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir,
                          save_prefix='cat', save_format='jpeg'):
    i += 1
    if i >= 20:  # 20 yeni resim üret
        break

print("20 yeni resim kaydedildi:", save_dir)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

# Resmi yükle
img = load_img("kon.jpg")
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

# Güvenli, hareketli augmentation
datagen = ImageDataGenerator(
    rotation_range=10,           # küçük dönüşler
    width_shift_range=0.03,      # yatay kaydırma az
    height_shift_range=0.03,     # dikey kaydırma az
    horizontal_flip=True,        # yatay çevirme
    brightness_range=(0.8, 1.2), # parlaklık
    channel_shift_range=20,      # renk kaydırma
    fill_mode='nearest'
)

# Klasör oluştur
save_dir = "augmented_images_movable"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Augmented resimleri kaydet
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir,
                          save_prefix='cat', save_format='jpeg'):
    i += 1
    if i >= 20:  # 20 resim üret
        break

print("20 güvenli ve hareketli resim oluşturuldu:", save_dir)
```

```python
import os
import shutil
import random

# Kaynak klasör (tüm resimlerin bulunduğu klasörün yolu)
source_dir = r"C:\Users\theni\OneDrive\Masaüstü\deneme"


# Hedef klasörler
train_dir = "train"
test_dir = "test"

# Klasörler yoksa oluştur
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Sadece .jpeg dosyaları al
all_files = [f for f in os.listdir(source_dir) if f.lower().endswith('.jpeg')]

# Dosyaları karıştır
random.shuffle(all_files)

# %70 - %30 oranında ayır
split_index = int(len(all_files) * 0.7)
train_files = all_files[:split_index]
test_files = all_files[split_index:]

# Train dosyalarını kopyala
for f in train_files:
    shutil.copy(os.path.join(source_dir, f), os.path.join(train_dir, f))

# Test dosyalarını kopyala
for f in test_files:
    shutil.copy(os.path.join(source_dir, f), os.path.join(test_dir, f))

print(f"Toplam: {len(all_files)} resim")
print(f"Train: {len(train_files)} resim → {train_dir}/")
print(f"Test: {len(test_files)} resim → {test_dir}/")
```
