```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np

# 1. Resmi ykle
img = image.load_img("cat.jpg", target_size=(224, 224))  # resmi 224x224'e yeniden boyutlandryoruz

# 2. NumPy array'e evir
img_array = image.img_to_array(img)

# 3. Normalizasyon (0-1 aralna getiriyoruz)
img_array = img_array / 255.0

# 4. Model giriine uygun ekle getir (batch ekle)
img_array = np.expand_dims(img_array, axis=0)

print("Resim array boyutu:", img_array.shape)

# 5. Grselletir
plt.imshow(img)
plt.title("Yklenen Kedi Resmi")
plt.axis("off")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

# 1. Tek resmimizi ykle
img = load_img("cat.jpg")       # resmin yolu
x = img_to_array(img)           # numpy array'e evir
x = x.reshape((1,) + x.shape)   # batch boyutu ekle

# 2. Data augmentation tanmla
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 3. Klasr olutur (k klasr)
save_dir = "augmented_images"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 4. Augmented grntleri ret ve kaydet
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir,
                          save_prefix='cat', save_format='jpeg'):
    i += 1
    if i >= 20:  # 20 yeni resim ret
        break

print("20 yeni resim kaydedildi:", save_dir)
```

```python
import os
import shutil
import random

# Kaynak klasr (tm resimlerin bulunduu klasrn yolu)
source_dir = r"C:\Users\theni\OneDrive\Masast\deneme"


# Hedef klasrler
train_dir = "train"
test_dir = "test"

# Klasrler yoksa olutur
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Sadece .jpeg dosyalar al
all_files = [f for f in os.listdir(source_dir) if f.lower().endswith('.jpeg')]

# Dosyalar kartr
random.shuffle(all_files)

# %70 - %30 orannda ayr
split_index = int(len(all_files) * 0.7)
train_files = all_files[:split_index]
test_files = all_files[split_index:]

# Train dosyalarn kopyala
for f in train_files:
    shutil.copy(os.path.join(source_dir, f), os.path.join(train_dir, f))

# Test dosyalarn kopyala
for f in test_files:
    shutil.copy(os.path.join(source_dir, f), os.path.join(test_dir, f))

print(f"Toplam: {len(all_files)} resim")
print(f"Train: {len(train_files)} resim  {train_dir}/")
print(f"Test: {len(test_files)} resim  {test_dir}/")
```
